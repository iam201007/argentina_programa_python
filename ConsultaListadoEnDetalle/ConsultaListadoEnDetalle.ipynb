```python
import pandas as pd
import io # Necesario para usar los archivos cargados en Colab con pandas.read_csv
from google.colab import files # Necesario para usar los archivos cargados en Colab con pandas.read_csv
```

```python
# Archivos Excel a utilizar y campos de union --> archivo con el detalle
# El archivo tiene que estar cargado previamente en Google Colab
archivo_izquierda_nombre = "Surtido.xlsx"
archivo_izquierda_filasdescartar = 5 # cantidad de filas a descartar antes de los encabezados
archivo_izquierda_titulos = 0 # número de fila, restando 1, que contiene los encabezados
```

```python
# Archivos Excel a utilizar y campos de union --> archivo con el listado a consultar
# El archivo tiene que estar cargado previamente en Google Colab
archivo_derecha_nombre = "Listado.xlsx"
archivo_derecha_filasdescartar = 0 # cantidad de filas a descartar antes de los encabezados
archivo_derecha_titulos = 0 # número de fila, restando 1, que contiene los encabezados
```

```python
# Nombre del campo común en ambos listados para realizar la unión
nombrecampounion = "articulo"
```

```python
# --- Importar el archivo archivo_izquierda_nombre en un dataframe ---

try:
    # skiprows=5 descarta las primeras 5 filas (índices 0 a 4)
    # header=0 establece que la nueva primera fila (la fila 6 original) se use como encabezado
    detalle_df = pd.read_excel(archivo_izquierda_nombre, skiprows=archivo_derecha_filasdescartar, header=archivo_izquierda_titulos)
    print(f"DataFrame cargado exitosamente con el contenido de {archivo_izquierda_nombre}.")
except Exception as e:
    print(f"Error al cargar 'pedido': {e}")
    pedido = pd.DataFrame() # Crear un DF vacío en caso de error para que el código no falle
```

```python
# --- Importar el archivo archivo_derecha_nombre en un dataframe ---

try:
    # skiprows=5 descarta las primeras 5 filas (índices 0 a 4)
    # header=0 establece que la nueva primera fila (la fila 6 original) se use como encabezado
    listado_df = pd.read_excel(archivo_derecha_nombre, skiprows=archivo_derecha_filasdescartar, header=archivo_derecha_titulos)
    print(f"DataFrame cargado exitosamente con el contenido de {archivo_derecha_nombre}.")
except Exception as e:
    print(f"Error al cargar 'pedido': {e}")
    pedido = pd.DataFrame() # Crear un DF vacío en caso de error para que el código no falle
```

```python
# realizar la unión entre el detalle y el listado para dejar solo los registros que coinciden ---

if not detalle_df.empty and not listado_df.empty:
    # La unión se hace por la columna indicada
    detallefiltrado_df = pd.merge(detalle_df, listado_df, on=nombrecampounion, how='inner')
    print("\nDataFrame 'detallefiltrado_df' (Inner Join) creado exitosamente.")
    print(f"Tamaño de 'detallefiltrado_df': {detallefiltrado_df.shape}")
else:
    detallefiltrado_df = pd.DataFrame()
    print("\nNo se pudo crear 'detallefiltrado_df' debido a un error en la carga de los DataFrames de origen.")
```

```python
# --- Exportación ---

output_filename = 'detallefiltrado_df.xlsx'
detallefiltrado_df.to_excel(output_filename, index=False)

print(f"\n--- PROCESO FINALIZADO ---")
print(f"Se ha generado el archivo: {output_filename}")
print(f"Filas encontradas: {len(detallefiltrado_df)}")

# Descarga automática
files.download(output_filename)
```
